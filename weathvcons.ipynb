```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file_path = 'raw_data/10yr_gas_data.csv'
# read in csv
gas_consump = pd.read_csv(file_path)

file_path2 = 'all_states_ave_temp.csv'
ave_temp = pd.read_csv(file_path2)
```

```python
#editing column names
gas_consump.head()
gas_consump = gas_consump.rename(columns={"Alabama Natural Gas Residential Consumption (MMcf)":"Alabama", "Alaska Natural Gas Residential Consumption (MMcf)":"Alaska", "Arizona Natural Gas Residential Consumption (MMcf)":"Arizona", "Arkansas Natural Gas Residential Consumption (MMcf)": "Arkansas", "California Natural Gas Residential Consumption (MMcf)":"California", "Colorado Natural Gas Residential Consumption (MMcf)":"Colorado", "Connecticut Natural Gas Residential Consumption (MMcf)":"Connecticut", "Delaware Natural Gas Residential Consumption (MMcf)":"Delaware","District of Columbia Natural Gas Residential Consumption (MMcf)":"District", "Florida Natural Gas Residential Consumption (MMcf)":"Florida", "Georgia Natural Gas Residential Consumption (MMcf)":"Georgia", "Hawaii Natural Gas Residential Consumption (MMcf)":"Hawaii", "Idaho Natural Gas Residential Consumption (MMcf)":"Idaho", "Illinois Natural Gas Residential Consumption (MMcf)":"Illinois", "Indiana Natural Gas Residential Consumption (MMcf)": "Indiana", "Iowa Natural Gas Residential Consumption (MMcf)": "Iowa"})
gas_consump
```

```python
gas_consump.head()
gas_consump.mean(axis=0)
# "Alabama Natural Gas Residential Consumption"].mean()
gasCon_mean_df = pd.DataFrame(gas_consump.mean(axis=0))
gasCon_mean_df
```

```python
gasCon_mean_df.reset_index(inplace=True)
gasCon_mean_df = gasCon_mean_df.rename(columns = {'': "States"})
gasCon_mean_df
```

```python
gasCon_mean_df = gasCon_mean_df.rename(columns = {'index': "States", 0:"Gas Consumption (MMcf)"})
gasCon_mean_df
```

```python
drop_DC_gasCon_mean_df = gasCon_mean_df.drop(8)
#clean_gasCon_mean_df = gasCon_mean_df.drop(11)
#clean_gasCon_mean_df = gasCon_mean_df.drop(51)
drop_hawaii_gasCon_mean_df = drop_DC_gasCon_mean_df.drop(11)
drop_US_gasCon_mean_df = drop_hawaii_gasCon_mean_df.drop(51)
statefiltered_gascon_mean_df = drop_US_gasCon_mean_df
#statefiltered_gascon_mean_df
```

```python
statefiltered_gascon_mean_df.reset_index(drop=True, inplace=True)
lenstates = len(statefiltered_gascon_mean_df["States"])

for i in range(lenstates):
    state = statefiltered_gascon_mean_df["States"][i]
    state = state.replace("Natural Gas Residential Consumption (MMcf)", "")
    # state_str = state_str.split(' Natural', 1)
    # state = state_str[0]
    statefiltered_gascon_mean_df["States"][i] = state
    print(state)
statefiltered_gascon_mean_df
```

```python
alabama_avetemp = ave_temp.loc[ave_temp["Unnamed: 1"]=="Alabama",:]

alabama_avetemp = pd.DataFrame(range(241), columns=['Average Temperature'])
alabama_avetemp= alabama_avetemp.astype("float")
```

```python
#Drop Nan values
ave_temp.dropna(inplace=True)
```

```python
ave_temp[" Average Temperature"]= ave_temp[" Average Temperature"].astype("float")
ave_temp.info()
grouped_ave_temp = ave_temp.groupby('Unnamed: 1', as_index=False)[' Average Temperature'].mean()

grouped_ave_temp = grouped_ave_temp.rename(columns = {"Unnamed: 1":"States"})
grouped_ave_temp
```

```python
ave_temp.mean(axis=0)
```

```python
#Join data

combined_data = statefiltered_gascon_mean_df.join(grouped_ave_temp, lsuffix="States")
combined_data
del combined_data["States"]
combined_data.rename(columns={"StatesStates":"States"}, inplace=True)
combined_data = combined_data[combined_data["Gas Consumption (MMcf)"]  >5000]

combined_data
```

```python
#ax = data_plot.add_subplot()
fig,ax1 = plt.subplots()
ax1.bar(combined_data['States'], combined_data["Gas Consumption (MMcf)"], color = 'blue', label="Average Gas Consumption (MMcf)")

ax1.set_xlabel("States")
ax1.set_ylabel("Gas Consumption (MMcf)")
plt.legend(loc='lower left', bbox_to_anchor=(1.1, 0.6))
ax2 = ax1.twinx()


ax2.set_ylabel("Temperature (Fahrenheit)")

ax2.plot(combined_data["States"], combined_data[" Average Temperature"], color="orange", label = " Average Temperature")
plt.legend(loc='lower left', bbox_to_anchor=(1.1, 0.5))
ax1.tick_params(axis='x', rotation=90)
plt.title("Average Gas Consumption vs Average Temperature per State")


# combined_data["Gas Consumption (MMcf)"].plot(kind='bar', color='blue', ax=ax1)
# combined_data[" Average Temperature"].plot(kind='line', color='orange', ax=ax2)

```
